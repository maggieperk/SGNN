```python
# SGNN expects data in the form of Text only so we want to drop the other columsn from the talkmoves transcripts
# TalkMoves transcripts in the public data set have these columns: TimeStamp	Turn	Speaker	Sentence	T Tag	Student Tag
```

```python
# Imports
import os
import pandas as pd
```

```python
def convert_transcript_to_sgnn_format(transcript_filepath, output_filepath):
    original_df = pd.read_excel(transcript_filepath, header=0, engine='openpyxl')
    text_only_df = original_df['Sentence'].to_frame().rename(columns={"Sentence": "Text"})
    text_only_df.to_csv(output_filepath)
    print(f"Sentences read {text_only_df.size}")
    
def convert_all_transcripts_to_sgnn_format(transcript_dir, output_dir):
    transcript_count = 0
    errors = 0
    for root, dirs, files in os.walk(transcript_dir):
        for file in files:
            if file.endswith('.xlsx'):
                try:
                    transcript_filepath = os.path.join(root, file)
                    new_filename = f"sgnn_format_{file}"
                    output_filepath = f"{output_dir}/{new_filename}"
                    convert_transcript_to_sgnn_format(transcript_filepath, output_filepath)
                    transcript_count +=1 
                except Exception:
                    errors +=1
    print(f"Files converted: {transcript_count}. Conversion errors: {errors}.")
```

```python
convert_transcript_to_sgnn_format('data/7th_grade_math.xlsx', 'data/sgnn_format_test.xlsx')
```
